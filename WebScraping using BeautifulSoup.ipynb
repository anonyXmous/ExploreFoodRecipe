```python
import pandas as pd
import numpy as np
```

```python
import urllib
from bs4 import BeautifulSoup
```

```python
main_page_url = "https://www.bigoven.com/recipes/main-dish/grill-and-bbq"
```

```python
main_page = urllib.request.urlopen(main_page_url)
soup = BeautifulSoup(main_page, 'lxml')
```

```python
recipe_dict = {}

for link in soup.find_all("div", attrs = {"class": "recipe-tile-full"}):
    # Get to each recipe page
    recipe_url = link.get('data-url')
    recipe_page = urllib.request.urlopen(recipe_url)
    recipe_soup = BeautifulSoup(recipe_page, 'lxml')
    
    # Take the recipe name
    recipe_name = recipe_soup.find("h1").text
        
    # Take the ingredients (name and amount)
    ingredients_table = recipe_soup.find("div", {"class": "ingredientbox"})
    ingredients_dict = {}
    for amt_class, name_class in zip(ingredients_table.find_all("span", {"class": "amount"}), ingredients_table.find_all("span", {"class": "name"})):
        amt, name = amt_class.text, name_class.text
        ingredients_dict[name] = amt
    
    # Include the recipe in recipe_dict
    recipe_dict[recipe_name] = ingredients_dict

recipe_dict
```
