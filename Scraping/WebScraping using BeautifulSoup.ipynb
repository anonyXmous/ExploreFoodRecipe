## Setup

```python
import pandas as pd
import numpy as np
```

```python
import urllib
from bs4 import BeautifulSoup
```

```python
main_page_url = "https://www.bigoven.com/recipes/main-dish/grill-and-bbq/page/"
```

## Create the list of all url for all recipes

```python
%%time
recipe_url_list = []
error_page_list = []

for i in range(1, 6):
    try: 
        with urllib.request.urlopen(main_page_url + str(i)) as url: 
            soup = BeautifulSoup(url, 'lxml')
                            
            # Get the url for all recipe page
            for link in soup.find_all("div", attrs = {"class": "recipe-tile-full"}):
                recipe_url = link.get('data-url') 
                recipe_url_list.append(recipe_url)
                
    except: 
        error_page_list.append(i)
```

```python
# Number of recipe pages? 
len(recipe_url_list)
```

```python
# which pages caused error?
error_page_list
```

## Scrape each recipe page

```python
%%time
recipe_dict = {}
error_recipe_list = []

for recipe_url in recipe_url_list: 

    try: 
        recipe_page = urllib.request.urlopen(recipe_url)
        recipe_soup = BeautifulSoup(recipe_page, 'lxml')

        # Take the recipe name
        recipe_name = recipe_soup.find("h1").text

        # Take the ingredients (name and amount)
        ingredients_table = recipe_soup.find("div", {"class": "ingredientbox"})
        ingredients_dict = {}
        for amt_class, name_class in zip(ingredients_table.find_all("span", {"class": "amount"}), ingredients_table.find_all("span", {"class": "name"})):
            amt, name = amt_class.text, name_class.text
            ingredients_dict[name] = amt

        # Include the recipe in recipe_dict
        recipe_dict[recipe_name] = ingredients_dict

    except: 
        error_recipe_list.append(recipe_url)
```

```python
# Check the scraped recipes
print("Number of recipes:", len(recipe_dict))
```
